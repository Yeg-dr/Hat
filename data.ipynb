```python
import numpy as np
```

```python
np.random.seed(42)
n_train = 10000

x_train = np.random.uniform(-1, 7, size=(n_train, 1))
y_train = np.random.uniform(-1, 9, size=(n_train, 1))
z_train = np.random.uniform(-10, 10, size=(n_train, 1))

X_train = np.hstack([x_train, y_train, z_train])
```

```python
np.random.seed(123)
n_test = 2000

x_test = np.random.uniform(-1, 7, size=(n_test, 1))
y_test = np.random.uniform(-1, 9, size=(n_test, 1))
z_test = np.random.uniform(-10, 10, size=(n_test, 1))

X_test = np.hstack([x_test, y_test, z_test])
```

```python
z_surf_train = 8.0 - (x_train - 3.0)**2 - (y_train - 5.0)**2
z_surf_test = 8.0 - (x_test - 3.0)**2 - (y_test - 5.0)**2
```

```python
Y_train = (z_train > z_surf_train). astype(np.int64)
Y_test = (z_test > z_surf_test).astype(np.int64)
```

```python
print("shapes: ", X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
for idx in [0, 1, 2, 3, 4]:
    print(idx, "x, y, z = ", X_train[idx], "z_surf = ", z_surf_train[idx], "label = ", Y_train[idx])

n_pos_train = np.sum(Y_train == 1)
n_neg_train = np.sum(Y_train == 0)
print("train: positives = ", n_pos_train, "negatives = ", n_neg_train)

n_pos_test = np.sum(Y_test == 1)
n_neg_test = np.sum(Y_test == 0)
print("train: positives = ", n_pos_test, "negatives = ", n_neg_test)
```

```python
assert not np.isnan(z_surf_train).any()
assert not np.isinf(z_surf_train).any()
assert Y_train.dtype == np.int64 or Y_train.dtype == np.int32
```

```python
np.save('Y_train.npy', Y_train)
np.save('X_train.npy', X_train)
np.save('Y_test.npy', Y_test)
np.save('X_test.npy', X_test)
```
